```python
import sklearn
import sklearn.datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
from sklearn.datasets import load_boston

#Loading boston dataset and printing the head
boston = load_boston()
df = pd.DataFrame(data = boston.data, columns = boston.feature_names)
print(df.head())
```

```python
df.hist(column=[ 'RM','AGE'],bins=15)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

features = df.drop('PRICE', axis = 1)
classe = df['PRICE']

l = LinearRegression()

l.fit(features, classe)

predict = l.predict(features)

#predicted vs actual
plt.scatter(classe, predict)
plt.xlabel("prices")
plt.ylabel("predicted prices")
plt.title("prices and predicted prices")

#score
print("Score: " + str(r2_score(classe, predict)))
```

```python
plt.figure('RM')
plt.scatter(df['RM'], classe)
plt.ylabel("Actual Values")
plt.xlabel('RM')
```
